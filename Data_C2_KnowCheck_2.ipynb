## Cleaning Data

Data taken from [Department of Veterans Affairs data](https://catalog.data.gov/organization/va-gov?res_format=CSV). It gives a good indication of where veteran service agency can concentration efforts. Data descriptionL "This report provides county-level estimates of the number of Veterans who received VA Disability Compensation benefits during fiscal year 2020. It includes the Veteransâ€™ total service-connected disability (SCD) rating, age group, and gender. Blank values represent small cell counts that have been suppressed to protect the identity of Veterans. In the "Total: Disability Compensation Recipients" column, each blank cell represents less than 10 Veterans. Some categories may not sum to the total due to missing information (e.g., age, gender, etc.)."

```python
import pandas as pd
import numpy as np
```

```python
vadc_df = pd.read_csv('https://raw.githubusercontent.com/shacktemp/knowledge_check_2/main/FY_2020_Disability_Compensation_Recipients_by_County.csv')
```

First minor issue that is going to be fixed is capitalizing all the words that are lowercase in the title heading.

```python
fixed_columns = {
    'FIPS code':'FIPS Code',
    'SCD rating: 0% to 20%':'SCD Rating: 0% to 20%',
    'SCD rating: 30% to 40%':'SCD Rating: 30% to 40%',
    'SCD rating: 50% to 60%':'SCD Rating: 50% to 60%',
    'SCD rating: 70% to 90%':'SCD Rating: 70% to 90%',
    'SCD rating: 100%':'SCD Rating: 100%',
    'Age: 65 or older':'Age: 65 or Older',

}

vadc_df.rename(columns=fixed_columns,inplace=True)
vadc_df.columns
```

The second clean performed will be to remove the rows with blanks in the "Total: Disability Compensation Recipients" cell because those rows do not have any information to analyze.



```python
vadc_df = vadc_df.dropna(subset=['Total: Disability Compensation Recipients'])
vadc_df.head(100)
```
